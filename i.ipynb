```python
import pandas as pd


ice_cream = {
    'flavour' : ['chocolate', 'vanilla', 'strawberry'],
    'price' : [2.5, 1.0, 1.5],
    'colour':['brown', 'white', 'pink'], 
    'stock' :[10, 20, 30]
}
df_icecream = pd.DataFrame(ice_cream)
df_icecream
```

```python
print("\n BASIC DATASET INFO.")
print(f"number of flavours : {len}(df_icecream)")
```

```python
df_icecream.columns
len(df_icecream.columns)
```

```python
total_price= df_icecream['price'].sum()
print(f"\n Total PRICE OF ALL FLAVOURS: ${total_price}")
```

```python
avarage_price = df_icecream['price'].mean()
print(f"\n AVARAGE PRICE OF ALL FLAVOURS: ${avarage_price}")
```

```python
median_price = df_icecream['price'].median()
print(f"\ MEDIAN PRICE OF ALL FLAVOURS: ${median_price}")
```

```python
min_price = df_icecream['price'].min()
print(f"\n MINIMUN PRICE OF ALL FLAVOURS: ${min_price}")
```

```python
max_price = df_icecream['price'].max()
print(f"\n MAXIMUN PRICE OF OF ALL FLAVOURS: ${max_price}")
```
